```python
import numpy as np

def get_rewards(dataset, theta, action_set):    
    if dataset == 'Steam':
        rewards = action_set @ theta
        min_r = np.min(rewards)
        max_r = np.max(rewards)
        rewards = (rewards - min_r) / (max_r - min_r)
    elif dataset == 'Movielens': 
        action_set, rewards = action_set
    return rewards
```

### Distribution of rewards for steam

```python
from datasets.load_data import *
import matplotlib.pyplot as plt

seed_data = 12
nb_runs = 50
dataset = 'Steam'

dfs_data = load_dataset(name=dataset, init=True)
for i in range(nb_runs):
    print('Seed: ', i)
    theta, action_set, nb_actions, d = load_dataset(df_data=dfs_data, name=dataset, seed=i)
    rewards = get_rewards(dataset, theta, action_set)
    plt.hist(rewards, bins=50)
    plt.xlabel('rating')
    plt.ylabel('frequence')
    plt.show()
```

### Distribution of rewards for movielens

```python
seed_data = 12
nb_runs = 50
dataset = 'Movielens'

dfs_data = load_dataset(name=dataset, init=True)
for i in range(nb_runs):
    print('Seed: ', i)
    theta, action_set, nb_actions, d = load_dataset(df_data=dfs_data, name=dataset, seed=i)
    rewards = get_rewards(dataset, theta, action_set)
    plt.hist(rewards, bins=50)
    plt.xlabel('rating')
    plt.ylabel('frequence')
    plt.show()
```
